## To fill in the null values in sixth month, we will extract the sixth month of 1998.

```python
sixth_month_df= df[(df['DATE']>'1998-06-01') & (df['DATE']<'1998-06-30')]
```

```python
sixth_month_df.dropna(inplace=True)
sixth_month_df
```

```python
PRCP_NotZero = sixth_month_df[sixth_month_df['PRCP']!=0]
PRCP_NotZero
```

- In the sixth month, we see if the total liquid water equivalent of presumably all precipitation (PRCP) is not zero, which means the rain status is true.

```python
PRCP_Zero = sixth_month_df[sixth_month_df['PRCP']==0]
PRCP_Zero
```

- In the sixth month, we see if the total liquid water equivalent of presumably all precipitation (PRCP) is zero, which means the rain status is false.

```python
zero = (PRCP_Zero['TMAX']-PRCP_Zero['TMIN']).median()
zero
```

```python
not_zero = (PRCP_NotZero['TMAX']-PRCP_NotZero['TMIN']).median()
not_zero
```

- Our approach to determining if PRCP is zero or not is the median of maxiumam temperature minus minumam temperature. Then see what the closest value is for the null values.

```python
def is_close(row,zero,not_zero):
    ZeroPRCP_median = 16.5
    NotZeroPRCP_median = 10.5
    
    Zero_Compare = abs(row['TMAX'] - row['TMIN']) - ZeroPRCP_median
    Not_ZeroCompare = abs(row['TMAX'] - row['TMIN']) - NotZeroPRCP_median
    
    if Zero_Compare < Not_ZeroCompare:
        return pd.Series({'PRCP': PRCP_Zero['PRCP'].median(), 'RAIN': False})
    else:
        return pd.Series({'PRCP': PRCP_NotZero['PRCP'].median(), 'RAIN': True})
```

```python
rows_with_nan = df[df['PRCP'].isna()][:2] # to take only null values in sixth month
filled_rows = rows_with_nan.apply(is_close,args=(zero,not_zero), axis=1)
filled_rows
```

```python
df.loc[rows_with_nan.index, ['PRCP', 'RAIN']] = filled_rows[['PRCP', 'RAIN']]
```

### Repeat the following steps for each column within outliers

* Visualize the outliers
* Detect if your data set contains any outliers
* Identify lower & higher limit of outliers
* Drop all outliers
* Visualize the column after processing


```python
plt.boxplot(df[['PRCP','TMAX','TMIN']],labels=['PRCP','TMAX','TMIN'])
plt.show()
```

# PRCP

```python
Q1 = np.percentile(df['PRCP'], 25, interpolation = 'midpoint')
Q3 = np.percentile(df['PRCP'], 75, interpolation = 'midpoint')
IQR = Q3 - Q1
               
print('lower bound outliers ', Q1 - 1.5*(IQR))
print('higher bound outliers', Q3 + 1.5*(IQR))
```

```python
df = df.drop(df[(df['PRCP'] > 0.25) | (df['PRCP'] < -0.15000000000000002)].index)
```

```python
plt.boxplot(df['PRCP'])
```
